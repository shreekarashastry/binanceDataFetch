```python
from bs4 import BeautifulSoup
```

```python
import codecs
f = codecs.open("text.html", 'r')
binance = f.read()
```

```python
soup = BeautifulSoup(binance, 'html.parser')
```

```python
soup.find(id="listing")
```

```python
coins = []
for each_Coin in soup.find_all('a'):
    coins.append(each_Coin.get_text())
    # coins.append(each_Coin.get('a'))
```

```python
coins.pop(0)
```

```python
coins
```

```python
#Let's get the zip files now
import requests

url = 'https://s3-ap-northeast-1.amazonaws.com/data.binance.vision?delimiter=/&prefix=data/spot/monthly/trades/'
main_url = 'https://data.binance.vision/'
```

```python
import zipfile
import io

for coin in coins:
    resp = requests.get(url+coin)

    coinMonthlyData = BeautifulSoup(resp.text, 'html.parser')
    allZipfiles = coinMonthlyData.find_all('key')
    
    for each in allZipfiles:
        if each.get_text().endswith('.zip'):
            r = requests.get(main_url+each.get_text())
            z = zipfile.ZipFile(io.BytesIO(r.content))
            z.extractall(coin)
            print(each.get_text())
```
